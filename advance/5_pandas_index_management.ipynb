```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#Load the dataset from hugging face
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#clean the data for date
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
    
```

```python
df_pakistan = df[df['job_country']=='Pakistan']
df_pakistan.reset_index(inplace=True)
df_pakistan
```

```python
median_pivot = df_pakistan.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc=['median','min','max'])
median_pivot
```

```python
median_pivot.sort_values(by=('median','salary_year_avg'))
```
